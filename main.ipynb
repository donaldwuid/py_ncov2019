# nCov2019数据分析

## 当前城市

当前显示的城市是：

```python
city_name = '深圳'
```

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import data_util
import plot_util
```

```python
raw_data = data_util.load_data()
```

## 累计数量

```python

city_daily_data = data_util.aggregate_daily(raw_data, city_name)
city_daily_data.tail(5)
```

```python
figure = plot_util.tsplot_conf_dead_cured(city_daily_data, city_name + '累计 ')
```

## 增长速度

```python
city_daily_data_1st_derivative = data_util.diff(city_daily_data)
city_daily_data_1st_derivative.tail()
```

```python
figure = plot_util.tsplot_conf_dead_cured(city_daily_data_1st_derivative, city_name + '速度 ')
```

## 增长加速度

```python
city_daily_data_2nd_derivative = data_util.diff(city_daily_data_1st_derivative)
city_daily_data_2nd_derivative.tail()
```

```python
figure = plot_util.tsplot_conf_dead_cured(city_daily_data_2nd_derivative, city_name + '加速度 ')
```
