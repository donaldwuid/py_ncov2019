# nCov2019数据分析

```python
city_name = '深圳'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import data_util
import plot_util
```

```python
raw_data = data_util.load_data()
raw_data.tail()
```

```python

city_daily_data = data_util.aggregate_daily(raw_data, city_name)
city_daily_data.tail()
```

```python
plot_util.tsplot_conf_dead_cured(city_daily_data, city_name + '累计 ')
```

```python
city_daily_data_1st_derivative = data_util.diff(city_daily_data)
city_daily_data_1st_derivative.tail()
```

```python
plot_util.tsplot_conf_dead_cured(city_daily_data_1st_derivative, city_name + '速度 ')
```

```python
city_daily_data_2nd_derivative = data_util.diff(city_daily_data_1st_derivative)
city_daily_data_2nd_derivative.tail()
```

```python
plot_util.tsplot_conf_dead_cured(city_daily_data_2nd_derivative, city_name + '加速度 ')
```
